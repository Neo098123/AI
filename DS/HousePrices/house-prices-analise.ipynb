# Análise dos dados

Para um melhor aproveitamento das informações no dataset, faremos uma breve interpretação dos dados

```python
#bibliotecas utilizadas:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df1 = df.drop(['id'], axis = 1)
sns.heatmap(df1.corr(), vmin = -1, vmax = 1)
df1.corr()
```

```python
df.corr()
```


O grafico de correlacao nos permite avaliar de forma rapida a relacao entre as features fornecidas,
uma cor mais clara ou mais escura significa uma correlacao mais forte entre a feature e o valor medio
 

```python
df.head()
```

```python
features = []
for x in df1:
    features.append(x)
del(features[13])
print(features)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.pairplot(df1, y_vars=['medv'],x_vars=features, height = 2.5)
```

É possível observar que as features crim, indus, rm e lstat seguem um padrão de funções contínuas, enquanto chas é binário, rad é inteiro e b aparece de maneira concentrada em situações de maior medv. as outras features seguem um padrão mais disperso em relação à medv.
